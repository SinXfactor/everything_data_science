# Matplotlib basics

### Importing Libraries along with inline matplotlib graph

```python
from matplotlib import pyplot as plt
import numpy as np

# one-time command to replace plt.show() everytime
%matplotlib inline  
```

```python
x = np.arange(0,10)
y = np.arange(10,20)
```

## Scatter plot
##### here: plt.scatter(x-data, y-data, color-code)

```python
plt.scatter(x,y,c='g');
```

##### xlabel, ylabel, plot-title

```python
plt.scatter(x,y,c='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph')
```

## Plot with lines and/or markers

```python
y = x*x
plt.plot(x,y);
```

##### here: plt.plot(x-data, y-data, line/marker style, line-thickness,marker-size)

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot');
```

## Subplots
##### here: subplot(rows, columns, position)

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r-+')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'c--');
```

### Function plots

```python
x = np.arange(1,11)
y = 2*x-1
plt.plot(x,y,'r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('func plot');
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title('sine-wave')
```

### Function subplots

```python
x = np.arange(0, 6*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sine-wave')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cos-wave');
```

## Bar Graph

```python
x = [2,5,3]
y = [4,7,3]
x2 = [4,10,6]
y2 = [2,3,1]
plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title('bar-graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis');
```

## Histogram

```python
a = np.random.randint(0,100,10)
print(f'data: {a}')
plt.hist(a)
plt.title('histogram');
```

## Boxplot
##### here: plt.boxplot(data, vertical/horizontal-box, box-fill)

```python
data = [np.random.normal(0,std,100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True);
```

## Pie-chart
##### here: plt.pie(data, slice-explode, label-to-give, slice-color, data-format, slice-shadow)

```python
labels = 'P', 'C', 'Js', 'Go'
sizes = np.random.randint(200,800,4)
colors = ['gold', 'yellowgreen', 'lightcoral', 'skyblue']
explode = (0.2, 0, 0, 0)  # here explode defines the distance(0.0-1.0) for every slice

plt.pie(sizes, explode = explode, labels = labels, colors = colors,
        autopct = '%1.1f%%', shadow = False)
plt.axis('equal');
```
