## ECDFplot (Emipircal Cumulative Distribution Function)

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.set_style('darkgrid')
```

```python
sns.ecdfplot(x='tip', data=tips)
```

```python
tips.tip.value_counts().head()
```

```python
sns.ecdfplot(x='tip', data=tips)
plt.axvline(4, c="black")
```

```python
sns.ecdfplot(y='tip', data=tips)
plt.axvline(0.5, c="black")
```

```python
tips.tip.median()
```

```python
sns.ecdfplot(x='tip', data=tips, hue='time')
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day')
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day', stat='count')
```

```python
sns.ecdfplot(x='tip', data=tips, complementary=True, stat='count')
```

```python
sns.ecdfplot(x='tip', data=tips, weights='tip')
plt.axhline(0.5, c="black")
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day',palette='summer', lw=3)
```

```python
p = sns.ecdfplot(x='tip', data=tips, hue='day',palette='summer', lw=3)
p.legend(['Thursday', 'Friday', 'Saturday', 'Sunday'], title='Day of week', loc=4)
```

### Violinplot: Options

```python
sns.violinplot(data=cars2[cars2.origin.isin(['japan', 'europe'])], x='cylinders', y='displacement', hue='origin')
plt.legend(loc=2)
```

```python
sns.violinplot(data=cars2[cars2.origin.isin(['japan', 'europe'])], x='cylinders', y='displacement', hue='origin', split=True)
plt.legend(loc=2)
```

```python
sns.violinplot(data=cars2[cars2.origin.isin(['japan', 'europe'])],
               x='cylinders', y='displacement', hue='origin',
              split=True, inner='quartiles')
plt.legend(loc=2)
```

```python
sns.violinplot(data=cars2[cars2.origin.isin(['japan', 'europe'])],
               x='cylinders', y='displacement', hue='origin',
              split=True, scale='count')
plt.legend(loc=2)
```

## ScatterPlot: is a Relational plot

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds
```

```python
diamonds2 = diamonds[diamonds.cut.isin(['Premium', 'Good']) & diamonds.color.isin(['D', 'F', 'J'])].sample(n=100, random_state=22)
diamonds2
```

```python
diamonds2.color.value_counts()
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds2)
```

```python
sns.scatterplot(x='x', y='y', data=diamonds2)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds2, hue='cut')
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds, hue='cut', palette=['purple', 'black'])
```
