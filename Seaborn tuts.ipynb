## KDEplot

```python
sns.set_style("darkgrid")
```

```python
sns.kdeplot(cars.horsepower)
```

```python
sns.kdeplot(cars.horsepower, shade=True)
```

```python
sns.kdeplot(cars.horsepower, bw_adjust=2)
```

```python
sns.kdeplot(cars.horsepower, cumulative=True)
```

### KDEplot: Bivariate

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=20)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=True, thresh=0.05 ,cbar=True)
```

## ECDFplot (Emipircal Cumulative Distribution Function)

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.set_style('darkgrid')
```

```python
sns.ecdfplot(x='tip', data=tips)
```

```python
tips.tip.value_counts().head()
```

```python
sns.ecdfplot(x='tip', data=tips)
plt.axvline(4, c="black")
```

```python
sns.ecdfplot(y='tip', data=tips)
plt.axvline(0.5, c="black")
```

```python
tips.tip.median()
```

```python
sns.ecdfplot(x='tip', data=tips, hue='time')
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day')
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day', stat='count')
```

```python
sns.ecdfplot(x='tip', data=tips, complementary=True, stat='count')
```

```python
sns.ecdfplot(x='tip', data=tips, weights='tip')
plt.axhline(0.5, c="black")
```

```python
sns.ecdfplot(x='tip', data=tips, hue='day',palette='summer', lw=3)
```

```python
p = sns.ecdfplot(x='tip', data=tips, hue='day',palette='summer', lw=3)
p.legend(['Thursday', 'Friday', 'Saturday', 'Sunday'], title='Day of week', loc=4)
```

## BoxPlot

```python
cars
```

```python
sns.set_style('whitegrid')
```

```python
cars2 = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.boxplot(x='mpg', data=cars2)
```

```python
cars2.mpg.describe()
```

```python
sns.boxplot(x='origin', y='mpg', data=cars2)
```

```python
sns.boxplot(x='origin', y='mpg', data=cars2, hue='cylinders')
```

```python
cars2.model_year.describe()
```

```python
cars2['newer_model'] = cars2.model_year > 76
```

```python
sns.boxplot(x='origin', y='mpg', data=cars2, hue='newer_model')
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2)
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, order=['japan', 'europe', 'usa'])
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, order=['japan', 'europe', 'usa'], hue='newer_model', hue_order=[True, False])
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, color='g')
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, order=['japan', 'europe', 'usa'], color='g', hue='newer_model')
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, width=0.5)
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, linewidth=2.5)
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, whis=1)
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, fliersize=10)
```

```python
sns.boxplot(y='origin', x='mpg', data=cars2, showcaps=False)
```
